```python
#Filtering Data Based on empty cells
import pandas as pd
df = pd.read_csv('data.csv')
new_df = df.dropna()
print(new_df.to_string())
```

```python
#Filtering based on specific condition
import pandas as pd
df = pd.read_csv('data.csv')
for x in df.index:
  if df.loc[x, "ChipRate"] > 15:
    df.drop(x, inplace = True)
print(df.to_string())
```

```python
#Cleaning the data by removing duplicxate data
import pandas as pd
df = pd.read_csv('data.csv')
df.drop_duplicates(inplace = True)
print(df.to_string())
```

```python
#Replacing missing values of a specific column
import pandas as pd
df = pd.read_csv('data.csv')
df["UCZAA"].fillna(1.2, inplace = True)
print(df.to_string())
```

```python
#Calculating the summary statistics for column ChipRate
import pandas as pd
df = pd.read_csv('data.csv')
x = df["ChipRate"].mean()
print(x)
y = df["ChipRate"].median()
print(y)
z = df["ChipRate"].mode()
print(z)
a = df["ChipRate"].std()
print(a)
```
